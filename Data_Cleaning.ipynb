### String to integer

```python
df['Unique ID'][0]
```

```python
df['Unique ID'] = [int(s[3:]) for s in df['Unique ID']]
df.head()
```

### Indexing

```python
df.set_index('Unique ID',inplace=True)
df.head()
```

### Categorical to Numeric

```python
df['Suburb'] = df['Suburb'].astype('category')
df['Suburb'] = df['Suburb'].cat.codes
df.head()
```

```python
df.dtypes
```

```python
df.drop(['Address'],axis=1,inplace=True)
df.head()
```

```python
df['Type'] = df['Type'].astype('category')
df['Type'] = df['Type'].cat.codes
df.head()
```

```python
df.dtypes
```

```python
df['CouncilArea'] = df['CouncilArea'].astype('category')
df['CouncilArea'] = df['CouncilArea'].cat.codes
df['Regionname'] = df['Regionname'].astype('category')
df['Regionname'] = df['Regionname'].cat.codes
df.head()
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df['Method'] = labelencoder.fit_transform(df['Method'])
df.head()
```

```python
df.dtypes
```

```python
labelencoder = LabelEncoder()
df['SellerG'] = labelencoder.fit_transform(df['SellerG'])
df.head()
```

```python
df.dtypes
```

### Nan Values

```python
df.isnull().sum()
```

```python
df.dropna(subset = ['Car'],inplace=True)
df.isnull().sum()
```

```python
df['BuildingArea'] = df['BuildingArea'].fillna(df['BuildingArea'].mean())
df.isnull().sum()
```

```python
df['YearBuilt'] = df['YearBuilt'].fillna(1970.0)
df.isnull().sum()
```

### Memory optimisation

```python
df['Car'] = df['Car'].astype('int')
df['YearBuilt'] = df['YearBuilt'].astype('int')
```
